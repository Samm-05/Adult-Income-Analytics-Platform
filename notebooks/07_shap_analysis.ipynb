```python
import shap
import pandas as pd
import joblib
```

```python
model = joblib.load(
    "../models/best_model.pkl"
)

df = pd.read_csv(
    "../data/processed/engineered_adult.csv"
)

X = df.drop(
    "income",
    axis=1
)
```

```python
explainer = shap.TreeExplainer(
    model
)

shap_values = explainer.shap_values(
    X
)
```
