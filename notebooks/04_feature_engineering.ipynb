```python
import pandas as pd

df = pd.read_csv(
    "../data/processed/cleaned_adult.csv"
)
```

```python
df["AgeGroup"] = pd.cut(
    df["age"],
    bins=[0,25,40,60,100],
    labels=[
        "Young",
        "Adult",
        "Middle",
        "Senior"
    ]
)
```

```python
df["HoursCategory"] = pd.cut(
    df["hours.per.week"],
    bins=[0,20,40,60,100],
    labels=[
        "Part-Time",
        "Standard",
        "Overtime",
        "Extreme"
    ]
)
```

```python
df["HasCapitalGain"] = (
    df["capital.gain"] > 0
).astype(int)
```

```python
df["HasCapitalLoss"] = (
    df["capital.loss"] > 0
).astype(int)
```

```python
df["income"] = df["income"].map({
    "<=50K":0,
    ">50K":1
})
```

```python
df = pd.get_dummies(
    df,
    drop_first=True
)
```
