Library Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Data

```python
df = pd.read_csv(
    "../data/processed/cleaned_adult.csv"
)
```

Income Distribution

```python
sns.countplot(
    data=df,
    x="income"
)

plt.show()
```

Age Distribution

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["age"],
    bins=30
)

plt.show()
```

Education Distribution

```python
plt.figure(figsize=(12,5))

sns.countplot(
    data=df,
    x="education"
)

plt.xticks(rotation=90)

plt.show()
```

Gender vs Income

```python
sns.countplot(
    data=df,
    x="sex",
    hue="income"
)

plt.show()
```

Occupation Analysis

```python
plt.figure(figsize=(15,6))

sns.countplot(
    data=df,
    y="occupation",
    order=df["occupation"]
    .value_counts()
    .index
)

plt.show()
```
