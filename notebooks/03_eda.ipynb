Library Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Data

```python
df = pd.read_csv(
    "../data/processed/cleaned_adult.csv"
)
```

Income Distribution

```python
sns.countplot(
    data=df,
    x="income"
)

plt.show()
```

Age Distribution

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["age"],
    bins=30
)

plt.show()
```
